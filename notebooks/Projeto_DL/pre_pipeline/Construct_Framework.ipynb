```
from google.colab import auth
from googleapiclient.http import MediaFileUpload
from googleapiclient.discovery import build

auth.authenticate_user()
drive_service = build('drive', 'v3')

def save_file_to_drive(name, path):
  file_metadata = {
    'name': name,
    'mimeType': 'application/octet-stream'
    }

  media = MediaFileUpload(path, 
    mimetype='application/octet-stream',
    resumable=True)

  created = drive_service.files().create(body=file_metadata,
                                  media_body=media,
                                  fields='id').execute()

  print('File ID: {}'.format(created.get('id')))

  return created
# save_file_to_drive(destination_name, path_to_file)
```

```
def check_raster_ok(filename,folder_path,raster_utils):
  file_path = os.path.join(folder_path,filename)
  raster = raster_utils.get_raster_infos(file_path)
  del raster
```

```
def check_numpy_ok(filename,folder_path,utils_methods):
  file_path = os.path.join(folder_path,filename)
  np_array = utils_methods.retrieve_data_from_np_array(file_path)
  del np_array
```

```
import pandas as pd
def check_csv_ok(filename,folder_path):
  file_path = os.path.join(folder_path,filename)
  csv_df = pd.read_csv(file_path)
  del csv_df
```

```
def check_txt_ok(filename,folder_path):
  file_path = os.path.join(folder_path,filename)
  f = open(file_path, "r")
  txt_file = f.read()
  del txt_file
```
